```sql
WITH cte_avg AS (
    SELECT TOP 5 FactSales.Quantity, DimProduct.ProductName
    FROM DimProduct
    JOIN FactSales ON FactSales.ProductCode = DimProduct.ProductCode
    ORDER BY FactSales.Quantity -- Ensuring consistent selection
),
cte_avg_calc AS (
    SELECT AVG(Quantity) AS avg_quantity FROM cte_avg
)
SELECT cte_avg.*
FROM cte_avg
JOIN cte_avg_calc ON cte_avg.Quantity > cte_avg_calc.avg_quantity;
```
